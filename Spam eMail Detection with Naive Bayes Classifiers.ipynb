```python
df=pd.read_csv('/content/emails.csv')
```

```python
df['spam'].value_counts()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
#separate x and y
x= df.text.values
y=df.spam.values
```

```python
#split dataset
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y, train_size=0.2)
```

```python
#Data Processing
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
x_train=cv.fit_transform(xtrain)
```

```python
xtrain.toarray()
```

```python
#ML Algorithm
#naive bayes used for classification problem,It is highly used in text classification.
#Types: Gaussian Naive Bayes classifier,Multinomial Naive Bayes,Bernoulli Naive Bayes
```

```python
#ay type data jonno multinomialNB besi effective
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(xtrain,ytrain)
```

```python
x_test=cv.transform(xtest)
```

```python
x_test.toarray()
```

```python
model.score(x_test,ytest)
```
